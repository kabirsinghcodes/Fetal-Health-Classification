# DATA VISUALIZATION

```python
# Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Ignoring warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the dataset

df=pd.read_csv('CTG.csv')
df.head()
```

```python
# Droping null values before data visulization

df.dropna(inplace=True)
df.isnull().sum()
```

### DATA VISUALISATION

```python
# Checking balanced dataset or not

pieplot=df['NSP'].value_counts()
ax=pieplot.plot.pie(figsize=(12,6), autopct='%1.2f%%', explode=[0.01,0.05,0.1], shadow=True)
```

THE DATASET IS IMBALANCED.....WE NEED TO BALANCE THE DATASET BEFORE USING ALGORITHMS

#### 1. Scatter plot

```python
sns.scatterplot(x=df['b'], y=df['e'],hue=df['NSP'])
plt.xlabel('Start Instant')
plt.ylabel('End Instant')
plt.title('Start and End istant with NSP')
```

#### 2. Histogram

```python
sns.histplot(x=df['ASTV'], hue=df['NSP'])
plt.title('percentage of time with abnormal short term variability VS NSP')
```

#### 3. Violin Plot

```python
sns.violinplot(data=df, x="NSP", y="A")
plt.ylabel('Calm sleep')
plt.title('Calm sleep with NSP')
```

#### 4. Count Plot

```python
sns.countplot(x=df['D'], hue=df['NSP'])
plt.title('Active violence vs NSP')
```

#### 5. Bar plot

```python
sns.barplot(x=df['C'], y=df['ALTV'])
plt.title('percentage of time with abnormal long term variability with Calm violence')
plt.xlabel('Calm Violence')
```

#### 6. Box plot

```python
plt.figure(figsize=(5,5))
sns.boxplot(df['LBE'])
plt.title('LBE')
plt.show()
```
