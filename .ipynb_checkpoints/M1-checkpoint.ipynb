# DATA ANALYSING AND PREPROCESSING

### DATA ANALYSING

```python
# Libraries

import pandas as pd
import numpy as np

# code to see all the columns

pd.set_option('display.max_columns', None)
```

```python
# Ignoring warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the dataset

df=pd.read_csv('CTG.csv')
df.head()
```

```python
# Reading the tail of the dataset

df.tail()
```

```python
# Total rows and columns in dataset

print('Total rows: {}'.format(df.shape[0]))
print('Total columns: {}'.format(df.shape[1]))
```

```python
# data types of each features

df.info()
```

```python
# Checking null values in each column

df.isnull().sum()
```

```python
# Total null values

sum(df.isnull().sum())
```

```python
# Checking duplicated records

duplicate=df.duplicated().sum()
print('Duplicated: {}'.format(duplicate))
```

```python
# Categorical columns vs Numerical columns

cat_columns=[feature for feature in df.columns if df[feature].dtype=='O']
num_columns=[feature for feature in df.columns if df[feature].dtype!='O']

print('Total categorical columns: {}'.format(len(cat_columns)))
print('Total numerical columns: {}'.format(len(num_columns)))
```

```python
# No of unique feature in categorical column

for i in cat_columns:
    print('{} : {}'.format(i,len(df[i].unique())))
```

```python
# Numerical columns

num_columns
```

```python
# Describe function for numerical columns

df.describe()
```

```python
# Correlation between the features

df.corr()
```

### DATA PREPROCESSING

```python
#### REMOVING NULL VALUES

df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Total null values

sum(df.isnull().sum())
```

```python
# Total rows and columns in dataset after removing null values

print('Total rows: {}'.format(df.shape[0]))
print('Total columns: {}'.format(df.shape[1]))
```

```python
# Checking the dataset is balanced or not

df['NSP'].value_counts()
```

Clearly we can tell the dataset is imbalanced
